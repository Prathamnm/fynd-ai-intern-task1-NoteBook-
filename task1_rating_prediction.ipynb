```python
def call_llm(prompt: str) -> str:
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt
    )
    return response.text.strip()
```

```python
def prompt_v1(review_text):
    prompt = f"""
    Predict the star rating (1 to 5) for the following Yelp review.
    Respond with ONLY a single integer.

    Review:
    {review_text}
    """
    return call_llm(prompt)
```

```python
preds_v1 = []

for text in df["review_text"]:
    try:
        preds_v1.append(int(prompt_v1(text)))
    except:
        preds_v1.append(None)

df["pred_v1"] = preds_v1
df.head()
```

```python
def prompt_v2(review_text: str) -> str:
    prompt = f"""
    You are an expert Yelp review analyst.

    Step 1: Analyze the sentiment (positive, neutral, negative).
    Step 2: Consider intensity of emotion and complaints/praise.
    Step 3: Predict the most appropriate Yelp star rating.

    IMPORTANT:
    - Output ONLY a single integer from 1 to 5.
    - Do NOT explain your reasoning.

    Review:
    {review_text}
    """
    return call_llm(prompt)
```

```python
preds_v2 = []

for text in df["review_text"]:
    try:
        preds_v2.append(int(prompt_v2(text)))
    except:
        preds_v2.append(None)

df["pred_v2"] = preds_v2
df.head()
```

```python
valid_v2 = df.dropna(subset=["pred_v2"])
accuracy_v2 = accuracy_score(valid_v2["actual_stars"], valid_v2["pred_v2"])
accuracy_v2
```

```python
def prompt_v3(review_text: str) -> str:
    prompt = f"""
    You are a Yelp rating prediction model.

    Examples:
    Review: "The food was amazing and service was excellent."
    Rating: 5

    Review: "It was okay, nothing special, but not bad."
    Rating: 3

    Review: "Terrible service, food was cold and rude staff."
    Rating: 1

    Now predict the Yelp rating (15) for the following review.
    Output ONLY the integer.

    Review:
    {review_text}
    """
    return call_llm(prompt)
```

```python
preds_v3 = []

for text in df["review_text"]:
    try:
        preds_v3.append(int(prompt_v3(text)))
    except:
        preds_v3.append(None)

df["pred_v3"] = preds_v3
df.head()
```

```python
valid_v3 = df.dropna(subset=["pred_v3"])
accuracy_v3 = accuracy_score(valid_v3["actual_stars"], valid_v3["pred_v3"])
accuracy_v3
```

```python
from sklearn.metrics import accuracy_score

# V1 accuracy
valid_v1 = df.dropna(subset=["pred_v1"])
accuracy_v1 = accuracy_score(valid_v1["actual_stars"], valid_v1["pred_v1"])

# V2 accuracy
valid_v2 = df.dropna(subset=["pred_v2"])
accuracy_v2 = accuracy_score(valid_v2["actual_stars"], valid_v2["pred_v2"])

# V3 accuracy
valid_v3 = df.dropna(subset=["pred_v3"])
accuracy_v3 = accuracy_score(valid_v3["actual_stars"], valid_v3["pred_v3"])

accuracy_v1, accuracy_v2, accuracy_v3
```

```python
import pandas as pd

results = pd.DataFrame({
    "Prompt Version": [
        "V1 (Baseline)",
        "V2 (Structured Reasoning)",
        "V3 (Few-shot Prompting)"
    ],
    "Accuracy": [
        accuracy_v1,
        accuracy_v2,
        accuracy_v3
    ]
})

results
```

## Results & Analysis

Three prompting strategies were evaluated for Yelp star rating prediction.

- **V1 (Baseline)** achieved reasonable accuracy, showing that LLMs can map sentiment to ratings without guidance.
- **V2 (Structured Reasoning)** underperformed due to variability in reasoning and numeric extraction.
- **V3 (Few-shot Prompting)** achieved the highest accuracy by anchoring the model with concrete examples.

The results demonstrate that prompt engineering, especially few-shot learning, can significantly improve LLM performance without model fine-tuning. The perfect accuracy in V3 may indicate prompt-induced overfitting on a small sample size.
